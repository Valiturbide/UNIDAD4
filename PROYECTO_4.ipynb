```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Valiturbide/UNIDAD4/refs/heads/main/dataset.csv")
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

#**PARTICIPACION PARENTAL**#

```python
df['participacion_parental'].unique()
```

```python
import scipy.stats as stats
nivel_de_significancia =0.05

#Realizar ANOVA
anova_result = stats.f_oneway(
      df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
      df[df['participacion_parental'] == 'medio']['puntaje_examen'],
      df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos de los tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

```python
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

#**RESULTADOS**#
>ANOVA:
Anova muestra que F-estadistico es de 80.46 lo cual indica la variacin de las medias de los grupos, por lo tanto se comprueba que hay diferencia significativa.

#**CONCLUSION**#
Los resultados arrojaron que hay diferencia signicativa entre al menos dos tratamientos, lo cual nos indica que efectivamente la participacin parental influye en el puntuaje del examen , mostrandonos que los padres que participan de manera elevada afectan positivamente en el resultado del puntuaje de sus hijos, por el contrario, los padres que tienen una participacin baja, afecta negativamente a sus hijos influyendo en su puntuaje y los que tiene una participacin media de igual manera las notas no son tan altas a diferencia de los padres que participan seguido, por lo tanto podra ser una rea de rforzamiento o mejora para que los padres participen ms en el desarrollo academico de sus hijos.
