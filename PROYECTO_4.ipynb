#**MARCO TEORICO**#

>ANOVA
>> Se define como  una técnica que permite comparar las varianzas entre las medias de diferentes muestras y sirve para analizar si existe alguna diferencia entre las medias de más de dos poblaciones. De manera que el análisis de la varianza permite determinar si las medias poblacionales de dos o más grupos son diferentes analizando la variabilidad entre las medias de las muestras.

>TUKEY
>>Se define como un método que tiene como fin comparar las medias individuales provenientes de un análisis de varianza de varias muestras sometidas a tratamientos distintos, además permite discernir si los resultados obtenidos son significativamente diferentes o no.

>VARIANZAS
>>Se define como una medida de dispersión que representa la variabilidad de una serie de datos respecto a su media.

>MEDIAS
>>Se define como el valor promedio de un conjunto de datos númericos,  Se calcula como la suma del conjunto de valores dividida entre el número total de valores.

>DISTRIBUCIÓN NORMAL
>>La distribución normal es simétrica, siendo la forma de la curva normal al lado izquierdo de la media, la imagen especular (el reflejo) de la forma al lado derecho de la media. Además se aplica de manera frecuente  para estudiar procesos cuya variable aleatoria es continua.

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/Valiturbide/UNIDAD4/refs/heads/main/dataset.csv")
df.drop(columns=[], inplace=True)
df.dropna(inplace=True)
df
```

#**PARTICIPACION PARENTAL**#

```python
df['participacion_parental'].unique()
```

```python
import scipy.stats as stats
nivel_de_significancia =0.05

#Realizar ANOVA
anova_result = stats.f_oneway(
      df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
      df[df['participacion_parental'] == 'medio']['puntaje_examen'],
      df[df['participacion_parental'] == 'elevado']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos de los tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

```python
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

#**RESULTADOS**#
>ANOVA:
Anova muestra que F-estadistico es de 80.46 lo cual indica la variación de las medias de los grupos, por lo tanto se comprueba que hay diferencia significativa.

#**CONCLUSION**#
Los resultados arrojaron que hay diferencia signicativa entre al menos dos tratamientos, lo cual nos indica que efectivamente la participación parental influye en el puntuaje del examen , mostrandonos que los padres que participan de manera elevada afectan positivamente en el resultado del puntuaje de sus hijos, por el contrario, los padres que tienen una participación baja, afecta negativamente a sus hijos influyendo en su puntuaje y los que tiene una participación media de igual manera las notas no son tan altas a diferencia de los padres que participan seguido, por lo tanto podría ser una área de rforzamiento o mejora para que los padres participen más en el desarrollo academico de sus hijos.
